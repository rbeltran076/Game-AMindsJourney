This is a game about freedom. The main character is a woman named Mina
 
Mina's mission is to get her constraints to 0. This will be the scoring system of the game.

Mina can eliminate a constraint by unlocking rooms, which will be arranged like this.

Mina spawns at the center of the map, she can go south to a lore section or north to a loot section.

She will need to investigate the lore section first so she can understand why she is there and how can she escape and be free.

In the lore section she will find a key to open one of the loot rooms and start collecting other keys to open the other 2 rooms
